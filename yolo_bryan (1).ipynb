```python
%cd /content/drive/MyDrive/FIT3163,3164/YOLO/dataset_3/
yaml_path = "data.yaml"

results = model.train(
    data=yaml_path,
    epochs=100,
    imgsz=640,
    batch=16,
    # lr0=0.0005,
    weight_decay=0.002,
    save_period=25,
    # device='cpu',

    # augmentations
    shear=0.15,             # Increase shear for geometrical robustness
    perspective=0.0005,     # Small perspective distortion
    hsv_h=0.2,             # Hue variation
    hsv_s=0.5,             # Saturation variation
    # flipud=0.1,
    close_mosaic=10 # stop mosaic augmentation in last 10 epochs
)
```
