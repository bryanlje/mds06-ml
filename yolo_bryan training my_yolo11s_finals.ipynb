```python
%cd /content/drive/MyDrive/FIT3163,3164/YOLO/dataset_4/
yaml_path = "data.yaml"

results = my_model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    # lr0=0.0005,
    weight_decay=0.005,
    dropout=0.4,
    save_period=25,
    # device='cpu',

    # augmentations
    shear=5,                # value: 0-180
    perspective=0.0005,     # Small perspective distortion, value=0-0.001
    hsv_h=0.5,              # Hue variation, value=0-1
    hsv_s=0.5,              # Saturation variation, value=0-1
    hsv_v=0.2,              # brightness, value=0-1
    fliplr=0.5,             # left-right flip
    close_mosaic=5,        # stop mosaic augmentation in last 10 epochs
    cutmix=0.2,

    # resume=True
)
```
